```python
a['date'] = a['date'].astype('datetime64')
```

```python
def sber_chart(name, region, date_min, date_max):
    a_modified = a[(a['date'].between(date_min, date_max)) & (a['name'] == name) 
 & (a['region'] == region)]
    a_modified.reset_index(drop=True, inplace=True)
    a_modified.plot(x='date', y='value')
    plt.xlabel('Период')
    plt.ylabel('Данные')
    plt.show()
                                      
sber_chart('Средняя пенсия', 'Москва', '2014-01-15', '2019-02-02')   
    
```
